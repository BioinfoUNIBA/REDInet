```python
# Copyright (c) 2023-2024 Pietro Luca Mazzacuva <pietroluca.mazzacuva@unicampus.it>
#
# Permission is hereby granted, free of charge, to any person obtaining a copy
# of this software and associated documentation files (the "Software"), to deal
# in the Software without restriction, including without limitation the rights
# to use, copy, modify, merge, publish, distribute, sublicense, and/or sell
# copies of the Software, and to permit persons to whom the Software is
# furnished to do so, subject to the following conditions:
#
# The above copyright notice and this permission notice shall be included in
# all copies or substantial portions of the Software.
#
# THE SOFTWARE IS PROVIDED "AS IS", WITHOUT WARRANTY OF ANY KIND, EXPRESS OR
# IMPLIED, INCLUDING BUT NOT LIMITED TO THE WARRANTIES OF MERCHANTABILITY,
# FITNESS FOR A PARTICULAR PURPOSE AND NONINFRINGEMENT. IN NO EVENT SHALL THE
# AUTHORS OR COPYRIGHT HOLDERS BE LIABLE FOR ANY CLAIM, DAMAGES OR OTHER
# LIABILITY, WHETHER IN AN ACTION OF CONTRACT, TORT OR OTHERWISE, ARISING FROM,
# OUT OF OR IN CONNECTION WITH THE SOFTWARE OR THE USE OR OTHER DEALINGS IN
# THE SOFTWARE.

import pandas as pd 
import numpy as np
import os, datetime
import logomaker as lm
import matplotlib.pyplot as plt
import seaborn as sn
from tqdm import tqdm
from datetime import datetime
from glob import glob

def counter(pandas_df, pos_index, snp_index, zero_index):
    
    indexes_list = [pos_index, snp_index, zero_index]
    counts = []
    for i in range(3):
        temp_counts = []
        for col in tqdm(range(101)):       
            a = pandas_df.iloc[indexes_list[i], col].value_counts()
            c = pandas_df.iloc[[j+1 for j in indexes_list[i]], col].value_counts()
            g = pandas_df.iloc[[j+2 for j in indexes_list[i]], col].value_counts()
            t = pandas_df.iloc[[j+3 for j in indexes_list[i]], col].value_counts() 

            if len(a)>0 and len(c)>0 and len(g)>0 and len(t)>0:
                try:
                    a_count = a[1]
                except:
                    if int(a.index.tolist()[0]) == 1.0:
                        a_count = a[0]
                    else:
                        a_count = 0.0
                try:
                    c_count = c[1]
                except:
                    if int(c.index.tolist()[0]) == 1.0:
                        c_count = c[0]
                    else:
                        c_count = 0.0
                try:
                    g_count = g[1]
                except:
                    if int(g.index.tolist()[0]) == 1.0:
                        g_count = g[0]
                    else:
                        g_count = 0.0
                try:
                    t_count = t[1]
                except:
                    if int(t.index.tolist()[0]) == 1.0:
                        t_count = t[0]
                    else:
                        t_count = 0.0

                temp_counts.append([a_count, c_count, g_count, t_count])
            else:
                temp_counts.append([0.0, 0.0, 0.0, 0.0])
        counts.append(temp_counts)
        
    return counts[0], counts[1], counts[2] 
    

base_inputdir_path = "/lustre/home/pietrolucamazzacuva"
tissues_path = "/lustrehome/pietrolucamazzacuva/filezilla-recas/tissues"

model_path = "/lustrehome/pietrolucamazzacuva/filezilla-recas/REDInet_models"
if not os.path.exists(model_path):
    os.mkdir(model_path)

model_name = f'REDInet_log_preprocessing_pos_zeros_snps_{datetime.now().strftime("%d_%m_%Y_%H_%M_%S")}.h5'

body_site_table = pd.read_table("/lustrehome/pietrolucamazzacuva/filezilla-recas/tissues/SRR_list.tsv")

features_total_filepaths = []
metadata_total_filepaths = []

for c,srr in enumerate(body_site_table["Sample"].tolist()):
    print(f"[{datetime.now()}] Loading SRR {srr} features vectors and metadata.")
    features_filepath = glob(os.path.join(base_inputdir_path, "*", f"{srr}_features_reduced_pos_zeros_snps.tsv"))
    metadata_filepath = glob(os.path.join(base_inputdir_path, "*", f"{srr}_metadata_reduced_pos_zeros_snps.tsv"))
    features_total_filepaths += features_filepath
    metadata_total_filepaths += metadata_filepath
    
    
print(f"Total processed SRR table:", c+1)
print(f"Len. features_total_filepaths:", len(features_total_filepaths))
print(f"Len. metadata_total_filepaths:", len(metadata_total_filepaths))

start = datetime.now()
X_list = []
with tqdm(total=len(features_total_filepaths)) as pbar:
    for features_filepath in features_total_filepaths:
        X_list.append(pd.read_table(features_filepath, header=None))
        pbar.update(1)
X = pd.concat(X_list, axis=0)
del(X_list)

y_list = []
with tqdm(total=len(metadata_total_filepaths)) as pbar:
    for metadata_filepath in metadata_total_filepaths:
        y_list.append(pd.read_table(metadata_filepath, sep="\t"))
        pbar.update(1)
y = pd.concat(y_list, axis=0)
del(y_list)
print(f"Elapsed time for dataset loading: {datetime.now()-start}")

X.reset_index(drop=True, inplace=True)
y.reset_index(drop=True, inplace=True)

a_index = [i*8 for i in range(len(y))]
zoom = X.iloc[a_index, 50]
del a_index
t_index = zoom[zoom.iloc[:]==0.0].index.tolist()
del zoom
indexes = []
for i in t_index:
    indexes += [i+j for j in range(8)]
del t_index
copy = X.iloc[indexes]
copy.reset_index(drop=True, inplace=True)
cols_reverse = list(X.columns)
cols_reverse.sort(reverse=True)
copy = copy.loc[:, cols_reverse]
del cols_reverse
index_rows_start = [i for i in range(0, len(copy)-7, 8)]
index_reverse = []
for row in index_rows_start:
    index_reverse += [row+3, row+2, row+2, row, row+7, row+6, row+5, row+4]
del index_rows_start
copy = copy.iloc[index_reverse, :]
del index_reverse
X.iloc[indexes, :] = copy

X = X.to_numpy()
X = np.reshape(X, (int(X.shape[0]/8), 8, 101))
X, indexes = np.unique(X, axis=0, return_index=True)
X = pd.DataFrame(np.reshape(X, (int(X.shape[0]*8), 101)))
y = y.iloc[indexes, :]
y.reset_index(drop=True, inplace=True)
               
p_index = y[y.loc[:, "Type"].str.find("POSITIVE")!=-1].index.tolist()
z_index = y[y.loc[:, "Type"].str.find("ZERO")!=-1].index.tolist()
s_index = y[(y.loc[:, "Type"].str.find("POSITIVE")==-1) & (y.loc[:, "Type"].str.find("ZERO")==-1)].index.tolist()

n_index = z_index + s_index
n_index.sort()

AG_N = X.iloc[[(i*8)+6 for i in n_index], 50].value_counts()
N_X = AG_N.index.tolist()
N_Y = AG_N.iloc[:].tolist()
N_Y = [i/sum(N_Y) for i in N_Y]

AG_P = X.iloc[[(i*8)+6 for i in p_index], 50].value_counts()
P_X = AG_P.index.tolist()
P_Y = AG_P.iloc[:].tolist()
P_Y = [i/sum(P_Y) for i in P_Y]

fig, ax = plt.subplots()

ax = sn.kdeplot(P_Y, x = P_X, color="dodgerblue")
ax = sn.kdeplot(N_Y, x = N_X, color="orangered")
plt.xlabel("AG Substitutions Rate ", fontsize=10)
plt.ylabel("Density", fontsize=10)
plt.legend(labels=['Editing Sites', 'Not Editing Sites'])
plt.title("Central Nucleotide AG Substitutions Rate KDE", fontsize=10.5) 
plt.savefig(os.path.join(tissues_path, "Dataset_Central_Nucleotide_AG_Rate_KDE.png")) 

plt.show()

del n_index, fig, ax, AG_N, N_X, N_Y, AG_P, P_X, P_Y  

indexes_p = []
for i in p_index:
    indexes_p.append(i*8)
    
indexes_s = []
for i in s_index:
    indexes_s.append(i*8)
    
indexes_z = []
for i in z_index:
    indexes_z.append(i*8)    
    
counts_pos, counts_snps, counts_zeros = counter(X, indexes_p, indexes_s, indexes_z)

del X, y

index = pd.Index([f"Pos {i}" if i <=0  else f"Pos +{i}" for i in range(-50, 51, 1)], name="Nucleotide Position")

counts_pos = pd.DataFrame(counts_pos)
counts_pos.columns = ["A", "C", "G", "T"]
counts_pos.set_index(index, inplace=True, drop=True)
counts_pos.to_csv(os.path.join(tissues_path, "positives_bases_counts_for_positions.tsv"), sep="\t")

counts_snps = pd.DataFrame(counts_snps)
counts_snps.columns = ["A", "C", "G", "T"]
counts_snps.set_index(index, inplace=True, drop=True)
counts_snps.to_csv(os.path.join(tissues_path, "snps_bases_counts_for_positions.tsv"), sep="\t")

counts_zeros = pd.DataFrame(counts_zeros)
counts_zeros.columns = ["A", "C", "G", "T"]
counts_zeros.set_index(index, inplace=True, drop=True)
counts_zeros.to_csv(os.path.join(tissues_path, "zeros_bases_counts_for_positions.tsv"), sep="\t")

counts_pos = counts_pos.divide(counts_pos.sum(axis=1), axis=0)
counts_snps = counts_pos.divide(counts_snps.sum(axis=1), axis=0)
counts_zeros = counts_zeros.divide(counts_zeros.sum(axis=1), axis=0)

counts_pos = counts_pos.sub(counts_snps.add(counts_zeros))
counts_pos.reset_index(drop=True, inplace=True)

%matplotlib inline

logo = lm.Logo(counts_pos.iloc[30:71, :], font_name = 'Arial Rounded MT Bold', figsize=(10, 3.75))
logo.ax.set_ylabel('Depletion            Enrichment', labelpad=0.5)
logo.ax.set_xticklabels('%+d'%x if x!=0 else '%d'%x for x in [-25, -20, -15, - 10, -5, 0, 5, 10, 15, 20])
logo.ax.set_xlabel('Nucleotide Position', labelpad=0.5)
logo.ax.set_title("Positives vs Negatives Enrichment-Depletion Logo")

plt.savefig(os.path.join(tissues_path, "Dataset_Logo.png"))
```
